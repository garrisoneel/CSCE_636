```python
import numpy as np
from collections import defaultdict
import re
words = {}
idxs = {}
freq = defaultdict(int)
total = 0
with open('./data/ptb/ptb.train.txt','r') as f:
    lines = f.readlines()

for line in lines:
    txt = re.findall(r'\S+', line)
    for word in txt:
        if word not in words:
            index = len(words)
            words[index] = word
            idxs[word] = index
        total+=1
        freq[word]+=1
```

```python
for F in freq:
    print(F, freq[F]/total)
```
